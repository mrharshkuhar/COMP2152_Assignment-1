```python
# Step 2: Dictionary
# Dictionary storing workout minutes for each friend: dict[str, tuple[int, int, int]]
workout_stats = {
    "Chris": (40, 35, 30),
    "Morgan": (25, 55, 45),
    "Casey": (50, 20, 40)
}

```

```python
# Step d: Cal# Calculate total workout minutes and add new key-value pairs
for name in list(workout_stats.keys()):
    total_time = sum(workout_stats[name])
    workout_stats[name + "_Total"] = total_timeculate total workout minutes using a loop and add to dictionary
```

```python
# Step 3: 2D List
# Nested list storing workout minutes for each friend: list[list[int]]
workout_list = []

for value in workout_stats.values():
    if isinstance(value, tuple):
        workout_list.append(list(value))d list called workout_list
```

```python
# Step f: Slice the workout_list
# Slice 1: Yoga and Running minutes for all friends
yoga_running = [activity[:2] for activity in workout_list]
print("Yoga and Running minutes for all friends:", yoga_running)

# Slice 2: Weightlifting minutes for last two friends
weightlifting_last_two = [activity[2] for activity in workout_list[-2:]]
print("Weightlifting minutes for last two friends:", weightlifting_last_two)
```

```python
 # Step 4: Check for total workout >= 120 minutes
for key in workout_stats:
    if "_Total" in key and workout_stats[key] >= 120:
        friend_name = key.replace("_Total", "")
        print(f"Great job staying active, {friend_name}!")total >= 120
```

```python
# Step 5: User Input Feature
friend_name = input("Enter friend's name: ")

if friend_name in workout_stats and isinstance(workout_stats[friend_name], tuple):
    activities = workout_stats[friend_name]
    total_minutes = sum(activities)
    print(f"\nWorkout details for {friend_name}:")
    print("Yoga:", activities[0], "minutes")
    print("Running:", activities[1], "minutes")
    print("Weightlifting:", activities[2], "minutes")
    print("Total:", total_minutes, "minutes")
else:
    print(f"Friend {friend_name} not found in the records.")
 up a friend
```

```python
# Step 6: Highest and Lowest Total Workout

totals = {}
for key, value in workout_stats.items():
    if "_Total" in key:
        totals[key] = value

highest_friend = max(totals, key=totals.get)
lowest_friend = min(totals, key=totals.get)

print("\nFriend with highest total workout minutes:",
      highest_friend.replace("_Total", ""),
      f"({totals[highest_friend]} minutes)")

print("Friend with lowest total workout minutes:",
      lowest_friend.replace("_Total", ""),
      f"({totals[lowest_friend]} minutes)")highest and lowest total workout minutes
```
